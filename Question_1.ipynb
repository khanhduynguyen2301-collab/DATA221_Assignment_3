```python
import pandas as pd

crime_csv_data = pd.read_csv("crime1.csv")

violent_crime = crime_csv_data[["ViolentCrimesPerPop"]]

mean = violent_crime["ViolentCrimesPerPop"].mean()
median = violent_crime["ViolentCrimesPerPop"].median()
std_dev = violent_crime["ViolentCrimesPerPop"].std()
minimum = violent_crime["ViolentCrimesPerPop"].min()
maximum = violent_crime["ViolentCrimesPerPop"].max()

print(f"The mean of the ViolentCrimesPerPop column is {mean}")
print(f"The median of the ViolentCrimesPerPop column is {median}")
print(f"The standard deviation of the ViolentCrimesPerPop column is {std_dev}")
print(f"The minimum value of the ViolentCrimesPerPop column is {minimum}")
print(f"The maximum value of the ViolentCrimesPerPop column is {maximum}")

# Looking at the mean and median, we notice that the mean is larger than the median (0.44119122257053295 > 0.39). Therefore, the distribution will be skewed, especially right-skewed. To explain this, we know a right-skewed distribution has a longer tail on the right side, which indicates that there is a few large values on the right tail and it pulls the mean upward. This results in the larger mean value compared to the median.

# If there are extreme values (very large or very small), the mean will be more affected than the median since the value will be significantly be change when there are extreme values as it computes the average. Meanwhile, the median just finds the value in the middle position; so if there are extreme values which lay on the sides, the median will barely move.
```
